## 1.1) Conheça as variáveis (*Features*) de seu conjunto de dados (*Dataset*)






Dicionário de Dados: https://archive.ics.uci.edu/ml/datasets/Heart+Disease 
<br />
*   **sex:** sexo do paciente [1-male; 0-female]
*   **cp (chest pain type):**  tipo de dor no peito [1-typical angina;  2-atypical angina; 3-non-angina pain; 4-asymptomatic]
*   **trestbps (resting blood pressure in mmHg on admission to the hospital)**: pressão arterial em repouso
*   **chol**: colesterol (em mg/dl)
*   **fbs:(fasting blood sugar>120 mg/dl)**: glicose sanguínea em jejum acima de 120 mg/dl [1-true; 2-false]
*   **restecg(resting electrocardiographic results)**: eletrocardiograma (ECG) em repouso [0-normal; 1-ST-T wave abnormality; 2-probable left ventricular hypertrophy]
*   **thalach (maximum heart rate achieved)**: batimento cardíaco máximo
*   **exang (exercise induced angina)**: angina induzida por exercício [1 = yes; 0 = no]
*   **oldpeak (ST depression induced by exercise relative to rest)**: depressão em segmento de onda induzido por exercício, durante ECG
*   **ca - number of major vessels (0-3)**: número de grandes vasos sanguíneos (ex: aorta, veia cava, artéria pulmonar) detectados em um exame de fluoroscopia
*   **thal**: talassemia (forma de anemia) [3 = normal; 6 = fixed defect; 7 = reversable defect] - De acordo com a época do estudo (1988)
*   **num**: **DESFECHO DO ESTUDO** (com ou sem doença cardiovascular) [0 = normal; 1-4 = heart disease]
*   **slope (slope of the peak exercise ST segment)**: inclinação de segmento ST durante pico de exercício [1 = upsloping; 2 = flat; 3 = downsloping] <br />

# 3) Recategorização de Variáveis 

Obs: Iremos fazer a recategorização, apenas da variável de desfecho (num). Em outro notebook, iremos preprocessar as variáveis para executar os modelos de *machine learning* (ML). Nestes casos, veremos outras técnicas para variáveis qualitativas com mais de 2 valores (ex: one-hot encoding)

A variável 'num' que é o nosso desfecho possui 4 valores. <br />
No entanto, para o estudo, somente nos interessa se há doença (1,2,3) ou não (0), portanto, iremos recategorizar essa variável

```
df.num.value_counts()
```

```
# Uso a biblioteca numpy (renomeada para np) para filtrar números maiores que 0
# Se o valor for maior que zero (1 a 4), preencho com 1, senão preenço com 0
df['target'] = np.where(df.num > 0, 1, 0)
df.head()
```

Após a recategorização, podemos excluir a variável num, pois foi recodificada na variável target

```
df = df.drop(['num'],axis=1)
df.head()
```

## 4.2) Diagrama de caixas (*boxplot*)
Gráfico que permite visualizar a simetria ou assimetria dos dados, assim como identificar visualmente os valores atípicos ou aberrantes (muito abaixo ou muito acíma do esperado). Estes valores são chamados de *outliers*.<br /><br />

As caixas representam os dados entre o primeiro e o terceiro quartil (não abordados neste notebook). Os *whiskers* (fios de bigode) são as linhas que delimitam os valores limítrofes superiores e inferiores. A linha dentro da caixa representa a mediana (valor que divide os dados simetricamente ao meio). 



```
# 1) Passo a largura e altura (em polegadas) como parâmetros ao método hist()
# 2) Desabilito a saída do texto "array([...])" com ;
df_num.boxplot(figsize=(11,7));
```

É possível observar que há vários círculos acima ou abaixo das caixas. Estes são os *outliers*. Em uma análise de dados, é preciso discutir se estes valores são válidos e deverão ser incluídos na análise. Descreva novamente os dados para visualizar os valores mínimo, máximo e se são possíveis.

## 4.3) Gráficos de Dispersão (*scatterplot*)

Como o próprio nome diz, permitem visualizar a dispersão dos dados. Em um *scatterplot* bidimensional, podemos avaliar visualmente se há uma relação linear entre uma variável e outro. Ex: idade e maior valor de batimento cardíaco (thalach).

```
df_num.plot.scatter(x='age',y='thalach',c='blue')
```

**Aparentemente** observa-se no gráfico acima uma diminuição do valor máximo com o passar dos anos. **Não dá pra afirmar que uma coisa causa a outra, mas já temos uma ideia inicial.**

### 4.3.1) Matriz de Dispersão
Gera um gráfico de dispersão para todas as variáveis

```
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

#Ajusto a qualidade da imagem (dots per inch - dpi)
fig=plt.figure(dpi= 300)

#ajusto o tamanho da imagem
plt.rcParams['figure.figsize'] = [10, 8] #Default [6.4, 4.8]

scatter_matrix(df_num)
plt.show()
```

**Como interpretar?** : Se os valores estão espalhos, e não percebe-se aumento ou diminuição, é possível que não há relacionamento entre as variáveis. (**uma análise posterior de correlação é necessária**). <br/><br/>

Se ao aumentar os valores de uma variável, os valores da outra variável também aumentarem, pode-se supor uma relação positiva (ex: age e trestbps). Se ao aumentar os valores de uma variável houver diminuição dos valores da outra, pode-se supor uma relação negativa ou inversa (ex: age e thalach).

## 4.4) Análise de correlação
É possível verificar se há correlação linear entre as variáveis (ex: correlação de pearson). Variáveis altamente correlacionadas podem prejudicar o desenvolvimento de modelos, sendo necessário excluir uma das duas envolvidas.

```
corr = df_num.corr().abs()
corr
```

### 4.4.1) Magnitude (força) da correlação

Correlação.png

Mapa de calor (*Heatmap*) das correlações

```
# Usaremos a biblioteca seaborn para elaboração do heatmap
import seaborn as sns
```

```
#heatmap das variáveis com altissima correlação
f, ax = plt.subplots(figsize=(20,10))
corr = df_num.corr()
mask = np.zeros_like(corr, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220, 10, as_cmap = True)
sns.heatmap(corr,
            mask = mask, 
            cmap = cmap,
            center = 0,
            square = True, 
            linewidths = .5, 
            cbar_kws = {"shrink": .4},
            annot=True,
            vmin=-1,
            vmax=1)
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=10)
plt.title('Heatmap', size=15)
plt.show()
```

Como podemos observar, não há alta correlação entre as variáveis. Encontramos apenas correlações fracas ou muito fracas.

## 5.3) Gráfico de setores (a.k.a Gráfico de Pizza)

pizza.jpg

Fonte: Unsplash (Créditos: Ivan Torres)

Para visualizar as frequências relativas de uma variável, às vezes podemos utilizar o gráfico de setores, que pode exibir proporcionalmente cada categoria.

```
df_cat['sex'].value_counts().plot.pie()
```

Perceba que a fatia do sexo masculino (1.0) é maior do que a do sexo feminino, mostrando que a nossa amostra tem um maior número de participantes do sexo masculino.

Agora vejamos as frequências relativas

```
# 1) Defina uma legenda para cada fatia [0-feminino; 1-masculino]
labels = ['masculino', 'feminino']

# 2) Crie um vetor para armazenar as frequências absolutas
tamanho_fatias = [df_cat['sex'].value_counts()[0],
                  df_cat['sex'].value_counts()[1]]

# 3) Informe se há um "destaque" em uma das fatias da pizza
destaque = [0, 0.2] 

# 4) Finalmente gero o gráfico
df_cat['sex'].value_counts().plot.pie(x=tamanho_fatias, labels=labels, autopct='%1.1f%%', shadow=True, explode=destaque)
```


# 6) Valores Ausentes

A biblioteca missingno permite uma análise visual de valores ausentes.

```
import missingno as msno
msno.matrix(df)
```

**Obs:** Note que não há pequenas falhas nas barras ca e thal, significando que existem valores ausentes (not a number - nan) nestas colunas

```
#Conto quantos valores ausentes em cada coluna 
df[['ca','thal']].isna().sum()
```

## 6.1) Imputação de valores ausentes
Ao invés de remover as observações que possuem valores ausentes, podemos usar técnicas de *imputação*, que consiste em incluir valores como média, moda, mediana. <br /><br />

Podemos, por exemplo, realizar imputação pela média ou mediana nas variáveis quantitativas e realizar a imputação pela moda (categoria que mais se repete) nas variáveis categóricas. Ao nos aprofundarmos, poderemos usar técnicas mais avançadas de imputação, como o uso do algoritmo K-NN para preenchimento de valores conforme as observações mais parecidas com a obsevação em análise.

```
#Visualizo TODAS as linhas com valores ausentes
df_nan = df[df.isna().any(axis=1)]
df_nan
```

```
#Vejo a moda para a variável ca
df.ca.mode()
```

```
#Confirmo, visualizando as frequências
df.ca.value_counts()
```

```
#Conto a qtd de valores ausentes (NaNs)
df['ca'].isna().sum()
```

```
#Preencho as variáveis ausentes com a moda (valor 0)
df['ca'].fillna(0, inplace=True)
```

```
#Visualizo TODAS as linhas com valores ausentes
df_nan = df[df.isna().any(axis=1)]
df_nan
```

Veja que não há mais valores ausentes na variável ca. <br /><br />
Faço o mesmo procedimento com a variável thal

```
#Como é uma variável categórica, primeiro consulta a moda
df.thal.mode()
```

```
#Preencho as variáveis ausentes com a moda (valor 3)
df['thal'].fillna(3, inplace=True)
```

```
#Visualizo novamente a matrix de missing
msno.matrix(df)
```

**Não há mais missing!**

```
#Visualizo os valores imputados nas linhas 87 e 266 na variável thal
df.iloc[[87,266]]
```

```
#Verifico numericamente se há mais colunas com valores ausentes
df.isna().sum()
```
